```python
Companies_L =  pd.read_csv("Companyinfo_2000-2021_WithL.csv")
ETF_DF = pd.read_csv("etfs_details_type_fund_flow.csv")
ETF_DF
```

```python
ETF_list = ETF_DF['Symbol'].tolist()
ETF_list
```

```python
#Removing undesired rows by tics 
Companies_L = Companies_L[~Companies_L.tic.isin(ETF_list)]
Companies_L 
#Removing undesired rows by string in conm
Strings = ['ETF','ETN','MUNI','TRUST','INCOME','ISHARES']
for item in Strings:
    Companies_L= Companies_L[~(Companies_L.conm.str.contains(item) & (~Companies_L.conm.str.contains('INC')))]
    Companies_L= Companies_L[~(Companies_L.conm.str.contains(item) & (~Companies_L.conm.str.contains('FD')))]
Companies_L
```
