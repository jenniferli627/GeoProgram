```python
con_move=[]
con_nmove=[]
for com in company:
    com_df = data[data.tic == com]
    com_df = com_df.drop(columns= ['gvkey','datadate','indfmt','consol','popsrc','datafmt',
                                  'curcd','costat','fyrc','fyear'])
    comn_list=[]
    comn_list= com_df.add1.unique().tolist()
    if len(comn_list) == 1:
        con_nmove.append(com)
    else:
        con_move.append(com)
    print(comn_list)
```
