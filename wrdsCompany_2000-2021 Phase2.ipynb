```python
import numpy as np
import pandas as pd
import geopandas
import geopy
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="zhechen.zhang2@gmail.com")
```

```python
Companies = pd.read_csv("Companyinfo_2000-2021.csv")
Companies = Companies.loc[(Companies['cyear'] >= 2000)&(Companies['fyrc'] != 0)]
Companies = Companies.drop_duplicates(subset='tic', keep="first")
Companies
```

```python
Location = pd.read_csv("Location_dataframe.csv")
Location
```

```python

Companies_Manual = Companies.merge(Location, on='tic')
Companies_Manual = Companies_Manual.loc[Companies_Manual['longitude']=='Manual']
Companies_Manual
```

```python

f = lambda x: x["add1"].split(", ")[0]
Companies_Manual["add1"] = Companies_Manual.apply(f, axis=1)
Companies_Manual
```

```python
latitude_Manual = []
longitude_Manual = []
for row in Companies_Manual['add1']:
    try:
        location = geolocator.geocode(row)
        latitude_Manual.append(location.latitude)
        longitude_Manual.append(location.longitude)
    except:
        latitude_Manual.append('Manual')
        longitude_Manual.append('Manual') 
```

```python
location_list_Manual= {'tic': ticker_Manual, 'longitude': longitude_Manual,'latitude':latitude_Manual}
Location_Manual = pd.DataFrame(data=location_list_Manual)
Location_Manual
```
