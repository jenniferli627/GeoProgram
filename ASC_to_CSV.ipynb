```python
for i in range(2000,2019):
    myArrayi  = np.loadtxt("V4NA03_PM25_NA_%d01_%d12-RH35-NoNegs.asc"%(i,i), skiprows=6)
    datai = pd.DataFrame(data=myArrayi)
    
    x_list=[]
    y_list=[]

    # all mines 0.005
    for xi in range(0,9300):
        x=round(xi*0.01-137.99,2)
        x_list.append(x)

    for yi in range(0,4550):
        y=round(yi*0.01+22.50,2)
        y_list.append(y)
        
    x_series = pd.Series(x_list)
    y_series = pd.Series(y_list)
    
    PMi = datai.rename(index=y_series, columns=x_series)
    
    print(PMi)
    
    PMi.to_csv("PM25_NA_%d.csv"%i)
```
